```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
temp['genre_id']= label_encoder.fit_transform(temp['genre'])
temp['genre_id'] += 1
temp['genre_id'].unique()
```

```python
temp['genre_id']
```

```python
# decision tree for feature importance on a classification problem
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot
# define dataset
cols=['year', 'sentence_count', 'char_count', 'unique words', 'word_count', 'unique word density','avg_word_length', 'expanded_word_count', 'contraction_difference', 'contraction_density']
a=temp[cols]
print("Original sape of input dataset",a.shape)                              #shape of original dataset 
b=temp['genre_id']
# define the model
model = DecisionTreeClassifier()
# fit the model
model.fit(a, b)
# get importance
importance = model.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
plt.xlabel("Imprtance wrt accuracy", labelpad=14)
plt.ylabel("HandCrafted Features", labelpad=14)
pyplot.show()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(a,b)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=a.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.xlabel("Imprtance wrt accuracy", labelpad=14)
plt.ylabel("HandCrafted Features", labelpad=14)
plt.show()
```

```python
df.fillna("",inplace=True)
df.isna().sum().sum()
```

```python
# extracting the required columns
X = df[['sentence_count', 'char_count', 'unique words', 'word_count', 'unique word density', 'avg_word_length', 'expanded_word_count', 'contraction_difference', 'contraction_density']]
Y = df['genre']

# splitting the data in train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15, random_state=42)
```

```python
X_train.shape
```

***Applying Logistic Regression***

```python
from sklearn.linear_model import LogisticRegression
```

```python
logReg = LogisticRegression(max_iter=500)
```

```python
logReg.fit(X_train, Y_train)
```

```python
Y_pred = logReg.predict(X_test)
```

```python
# Coefficients 
j = 0
for i in logReg.coef_ :
  print(logReg.classes_[j],i)
  j+=1
```

```python
print('Accuracy of logistic regression classifier on test set: {:.4f}'.format(logReg.score(X_test, Y_test)))
```

```python
from sklearn import metrics
```

```python
#Confusion Matrix
labels = logReg.classes_
conf_df = pd.DataFrame(metrics.confusion_matrix(Y_test, Y_pred), columns=labels, index=labels)
conf_df.index.name = 'True labels'
conf_df
```

```python
# Classification report : logistic regression 
print(metrics.classification_report(Y_test, Y_pred, target_names=logReg.classes_))
```

***Multinomial Naive Bayes***

```python
# importing required libraries
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
# fitting the train data
classifier = MultinomialNB().fit(X_train, Y_train)
```

```python
# predicting using test data
Y_pred = classifier.predict(X_test)
```

```python
# accuracy
print('Accuracy of Mutlinomial Naive Bayes classifier on test set: {:.4f}'.format(metrics.accuracy_score(Y_test, Y_pred)))
```

```python
#Confusion Matrix
labels = classifier.classes_
conf_df = pd.DataFrame(metrics.confusion_matrix(Y_test, Y_pred), columns=labels, index=labels)
conf_df.index.name = 'True labels'
conf_df
```

```python
# Classification report of the multinomial naive bayes classifier
print(metrics.classification_report(Y_test, Y_pred, target_names=classifier.classes_))
```

***Decision Tree***
