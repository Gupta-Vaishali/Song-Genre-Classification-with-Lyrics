```python
# Path of dataset
path = "/content/drive/MyDrive/lyrics.csv"
```

```python
# Importing required modules
import numpy as np
import pandas as pd
```

```python
# Loading csv into df
df=pd.read_csv(path)             
df.head()
```

```python
# Deleting column index
del df['index']                 
df.head()
```

```python
#Original no. of records in dataset
print(len(df))
```

```python
# Converting lyrics to lowercase
df['lyrics'] = df['lyrics'].str.lower()
df['lyrics'] = df['lyrics'].str.replace("\n"," ")
df['lyrics'] = df['lyrics'].str.replace("\t"," ")
df.head()
```

```python
# Removing songs with no lyrics
df.dropna(subset = ["lyrics"], inplace=True)
```

```python
# Now no. of records in dataset
print(len(df))
```

```python
# unique genres
genre_unique = df['genre'].unique()
print("\nTotal no. of unique genres present in dataset :",len(genre_unique))
for i in range(len(genre_unique)):
  if i != len(genre_unique) - 1 :
    print(genre_unique[i],",",end=" ")
  else :
    print(genre_unique[i]) 
```

```python
# Get names of indexes for which column genre has value not available
row_indices = df[ df['genre'] == 'Not Available' ].index
# Delete these row indexes from dataFrame
df.drop(row_indices , inplace=True)
```

```python
# Now no. of records in dataset
print(len(df))
```

```python
# Finding no. of unique song titles - maybe songs might be repeated
len(df['song'].unique())
```

```python
# Deleting records of songs that are repeated & printing no. of records in the dataset
df.drop_duplicates(subset=["song","artist",'genre',"lyrics"],keep='first',inplace=True)
print(len(df))
```

```python
# Deleting records of songs that are versions of same song but different artists & printing no. of records in the dataset
df.drop_duplicates(subset=["song",'genre',"lyrics"],keep='first',inplace=True)
print(len(df))
```

```python
# Deleting records of songs that are same just title has been changed & printing no. of records in the dataset
df.drop_duplicates(subset=["artist",'genre',"lyrics"],keep='first',inplace=True)
print(len(df))
```

```python
# Deleting records of songs that are same (genre and lyrics) & printing no. of records in the dataset
df.drop_duplicates(subset=['genre',"lyrics"],keep='first',inplace=True)
print(len(df))
```
